```python
import pickle
import matplotlib.pyplot as plt

def plot_metrics(filename='metrics.pkl'):
    """Load metrics from a pickle file and plot them."""
    with open(filename, 'rb') as f:
        metrics = pickle.load(f)
    
    print(filename)
    print(metrics)
    
    losses = metrics['losses']
    #losses = metrics['td_error']
    rewards = metrics['rewards']
    epsilon_history = metrics['epsilon_history']
    
    plt.figure(figsize=(12, 8))

    # Plot loss
    plt.subplot(3, 1, 1)
    plt.plot(losses)
    plt.ylim(0, 50)  # Set the max y-axis limit to 30
    plt.title('Average Loss Per Update')
    plt.xlabel('Episode')
    plt.ylabel('Average Loss')

    # Plot rewards (assuming rewards are collected per episode)
    plt.subplot(3, 1, 2)
    plt.plot(rewards)
    plt.title('Average Reward Per Update')
    plt.xlabel('Episode')
    plt.ylabel('Average Reward')

    # Plot epsilon decay
    #plt.subplot(3, 1, 3)
    #plt.plot(epsilon_history)
    #plt.title('Epsilon Decay')
    #plt.xlabel('Episode')
    #plt.ylabel('Epsilon')

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_metrics('agent_code/' + agent + '/metrics.pkl')
```
