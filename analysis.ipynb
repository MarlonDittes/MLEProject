```python
import pickle
import matplotlib.pyplot as plt

agent = 'QL_crate'

def plot_metrics(filename='metrics.pkl'):
    """Load metrics from a pickle file and plot them."""
    with open(filename, 'rb') as f:
        metrics = pickle.load(f)
    
    losses = metrics['td_error']
    rewards = metrics['rewards']
    epsilon_history = metrics['epsilon_history']
    
    plt.figure(figsize=(12, 8))

    # Plot loss
    plt.subplot(3, 1, 1)
    plt.plot(losses)
    plt.title('Average TD Error Per Episode')
    plt.xlabel('Episode')
    plt.ylabel('TD Error')

    # Plot rewards (assuming rewards are collected per episode)
    plt.subplot(3, 1, 2)
    plt.plot(rewards)
    plt.title('Average Reward Per Episode')
    plt.xlabel('Episode')
    plt.ylabel('Average Reward')

    # Plot epsilon decay
    plt.subplot(3, 1, 3)
    plt.plot(epsilon_history)
    plt.title('Epsilon Decay')
    plt.xlabel('Episode')
    plt.ylabel('Epsilon')

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_metrics('agent_code/' + agent + '/metrics.pkl')
```
